# Creamos el Dataframe de Gramaje

```python
import psycopg2
import pandas as pd

# Parámetros de conexión
params = {
    'dbname': 'acuarios2',
    'user': 'postgres',
    'password': 'sebacapo039',
    'host': 'localhost',
    'port': 5433
}

# Conectar a la base de datos
conn = psycopg2.connect(**params)

# Leer el archivo SQL
with open('./Listado_gramaje.sql', 'r') as file:
    sql_query = file.read()

# Ejecutar el archivo SQL y obtener los resultados
with conn.cursor() as cursor:
    cursor.execute(sql_query)
    # Suponiendo que es una consulta SELECT, recuperamos los datos
    records = cursor.fetchall()
    # Podemos convertir los resultados a un DataFrame de pandas para un manejo más fácil
    df = pd.DataFrame(records, columns=[desc[0] for desc in cursor.description])

# Cerrar la conexión
conn.close()

# Ahora df contiene la tabla de resultados y puedes trabajar con ella en Python
print(df)
```

```python
# Función para limpiar y reformatear los nombres de las columnas
def clean_column_names(columns):
    new_columns = []
    for col in columns:
        # Quitar el prefijo 'gra_' si existe
        if col.startswith('gra_'):
            col = col[4:]
        # Dividir por espacios y unir con '_'
        col = '_'.join(col.split())
        # Convertir a mayúsculas la primera letra
        col = col.capitalize()
        new_columns.append(col)
    return new_columns

# Aplicar la función para limpiar los nombres de las columnas
df.columns = clean_column_names(df.columns)
```

```python
# Convertir valores None a NaN
df['Peso_Promedio'] = df['Peso_Promedio'].astype(float)

# Calcular el incremento diario
df['Incremento'] = df['Peso_Promedio'].diff()

# Calcular el incremento semanal (7 días antes)
df['Incremento semanal'] = df['Peso_Promedio'] - df['Peso_Promedio'].shift(7)

df.head(10)
```

# Filtramos el Dataframe por solo una Piscina

```python
# Filtrar para obtener solo las filas donde la columna "Piscina" es igual a "01"
df = df[df['Piscina'] == '01']
```

```python
df['Fecha'] = pd.to_datetime(df['Fecha'] + ' 2023', format='%d %b %Y')
```

```python
# Interpolate missing values in 'Peso_Promedio', 'Sobrevivencia', 'Biomasa', 'Biomasa_Hectarea'
df['Peso_Promedio'] = df['Peso_Promedio'].interpolate(method='linear')
df['Sobrevivencia'] = df['Sobrevivencia'].interpolate(method='linear')
df['Biomasa'] = df['Biomasa'].interpolate(method='linear')
df['Biomasa_Hectarea'] = df['Biomasa_Hectarea'].interpolate(method='linear')
# Recreate 'Incremento' column. The increment is the difference in 'Peso_Promedio' compared to the previous day
df['Incremento'] = df['Peso_Promedio'].diff()
# Recreate 'IPS' column. The IPS (Indice Peso Semanal) is the weekly average of 'Peso_Promedio'
df['IPS'] = df['Peso_Promedio'].rolling(window=7).mean()
# Calculate 'CFR' column as the ratio between 'Balanceado' and 'Peso_Promedio'
df['Balanceado'] = df['Balanceado'].astype(float)
df['FCA'] = df['Balanceado'] / df['Peso_Promedio']

# Fill NaN values at the start of the columns with the first non-NaN value
df['Peso_Promedio'].fillna(method='bfill', inplace=True)
df['Sobrevivencia'].fillna(method='bfill', inplace=True)
df['Biomasa'].fillna(method='bfill', inplace=True)
df['Biomasa_Hectarea'].fillna(method='bfill', inplace=True)

# Recalculate 'Incremento', 'IPS', and 'CFR' columns
df['Incremento'] = df['Peso_Promedio'].diff()
df['IPS'] = df['Peso_Promedio'].rolling(window=7).mean()
```

```python
df.rename(columns={'Incremento semanal': 'Incremento semanal(promedio)'}, inplace=True)
```
